```python
from __future__ import print_function
from __future__ import division

import pandas as pd
import numpy as np

from glob import glob
from itertools import cycle

from skimage.io import imread, imread_collection 
from skimage.transform import rotate, resize, rescale
```

## Train

```python
dfs = []
for dir_path in glob('data/train/*'):
    label_name = dir_path.split('/')[-1]
    image_path = glob(dir_path + '/*.jpg')
    df_ = pd.DataFrame({ 'image_path': image_path, 'label_name': label_name })
    dfs.append(df_)
train_df = pd.concat(dfs, ignore_index=True)
```

```python
train_df.to_csv('data/train.csv')
```

## Test

```python
submission_sample_df = pd.read_csv('data/sampleSubmission.csv')
submission_sample_df.head()
```

```python
image_path = ['data/test/%s' % fname for fname in submission_sample_df['image'].values]
fname = submission_sample_df['image'].values
```

## Load to .npy

```python
h, w = 64, 64
```

```python
imgcol = imread_collection(train_df['image_path'], conserve_memory=True)
%time X_train = np.array([resize(img, [h, w]) for img in imgcol])
y_train = train_df['label_name']
np.save('data/X_train_%s' % h, X_train)
np.save('data/y_train_%s' % h, y_train)
```

```python
imgcol = imread_collection(image_path, conserve_memory=True)
%time X_test = np.array([resize(img, [h, w]) for img in imgcol])
np.save('data/X_test_%s' % h, X_test)
np.save('data/X_test_fname_%s' % h, fname)
```
