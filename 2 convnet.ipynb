## Convnet

```python
hw = 60
%time X_orig, y_orig = data.load_data(resize_shape=(hw, hw), scale=True)

enc = LabelEncoder()
X = X_orig.astype(np.float32)
y = enc.fit_transform(y_orig).astype(np.int32)

X_train, X_test, y_train, y_test = train_test_split(X.reshape(-1, hw * hw), y, test_size=0.05, random_state=42)
X_train, X_test = X_train.reshape(-1, 1, hw, hw), X_test.reshape(-1, 1, hw, hw) # because train_test_split requires array to X

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
if 'net' in locals():
    last_net = net

net = NeuralNet(
    eval_size=0.1,

    layers=[
        ('input', layers.InputLayer),

        ('conv1', layers.Conv2DCCLayer),
        ('pool1', layers.MaxPool2DCCLayer),
        ('dropout1', layers.DropoutLayer),

        ('conv2', layers.Conv2DCCLayer),
        ('pool2', layers.MaxPool2DCCLayer),
        ('dropout2', layers.DropoutLayer),

        ('conv3', layers.Conv2DCCLayer),
        ('pool3', layers.MaxPool2DCCLayer),
        ('dropout3', layers.DropoutLayer),

#         ('conv4', layers.Conv2DCCLayer),
#         ('pool4', layers.MaxPool2DCCLayer),

        ('hidden5', layers.DenseLayer),
#         ('maxpool5', layers.FeaturePoolLayer),
        ('dropout5', layers.DropoutLayer),

#         ('hidden6', layers.DenseLayer),
#         ('maxpool6', layers.FeaturePoolLayer),
#         ('dropout6', layers.DropoutLayer),

        ('output', layers.DenseLayer),
    ],

    input_shape=(None, 1, hw, hw),

    conv1_num_filters=32, conv1_filter_size=(5, 5),
    pool1_ds=(2, 2),
    dropout1_p=0.2,

    conv2_num_filters=64, conv2_filter_size=(5, 5),
    pool2_ds=(2, 2),
    dropout2_p=0.2,

    conv3_num_filters=128, conv3_filter_size=(3, 3),
    pool3_ds=(2, 2),
    dropout3_p=0.5,

#     conv4_num_filters=128, conv4_filter_size=(3, 3),
#     pool4_ds=(2, 2),
#     dropout4_p=0.5,

    hidden5_num_units=2048,
#     maxpool5_ds=2,
    dropout5_p=0.5,

#     hidden6_num_units=2048,
#     maxpool6_ds=2,

    output_num_units=len(set(y)),

    loss=negative_log_likelihood,

    batch_iterator_train=lasangeutils.MyBatchIterator(batch_size=512),
    batch_iterator_test=lasangeutils.MyBatchIterator(batch_size=512),

    update=nesterov_momentum,
    update_learning_rate=theano.shared(lasangeutils.float32(0.003)),
    update_momentum=theano.shared(lasangeutils.float32(0.9)),

#     update=rmsprop,
#     update=adadelta,

    regression=False,
    output_nonlinearity=softmax,

    on_epoch_finished=[
        lasangeutils.AdjustVariable('update_learning_rate', start=0.003, stop=0.001),
        lasangeutils.AdjustVariable('update_momentum', start=0.9, stop=0.999),
        lasangeutils.EarlyStopping(patience=50)
    ],

    max_epochs=1000,
    verbose=1
)

%time net.fit(X_train, y_train)
ascore = net.score(X_test, y_test)
# print(ascore)
```

```python
utils.plot_loss(net)
```

```python
y_pred = net.predict(X_test)
y_pred_proba = net.predict_proba(X_test)
lscore = utils.multiclass_log_loss(y_test, y_pred_proba)
ascore = net.score(X_test, y_test)
print('multiclass log loss', lscore)
print('accuracy', ascore)
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=False,  fmt='');
```

```python
utils.save(net, 'net-%s' % lscore)
```
