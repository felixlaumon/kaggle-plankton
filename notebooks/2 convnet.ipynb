## Convnet

```python
h, w = 64, 64

%time X_orig, y_orig = np.load('data/X_train_%s.npy' % h), np.load('data/y_train_%s.npy' % h)

X = 1 - X_orig
X = X.astype(np.float32)

enc = LabelEncoder()
y = enc.fit_transform(y_orig).astype(np.int32)

# X_train, X_test, y_train, y_test = train_test_split(X.reshape(-1, 64*64), y, test_size=0.25, random_state=42)
X_train, X_test, y_train, y_test = skextra.straified_train_test_split(X.reshape(-1, h*w), y, y=y, test_size=0.05, random_state=42)

X_train = X_train.reshape(-1, 1, h, w)
X_test = X_test.reshape(-1, 1, h, w)
X = X.reshape(-1, 1, h, w)

# Mean subtraction
# X_train_mean = np.mean(X_train, axis = 0)
# X_train -= X_train_mean
# X_test -= X_train_mean
# X_train_mean_reshaped = X_train_mean.reshape(h, w)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
n_class = len(enc.classes_)
classes = enc.classes_
plt.figure(figsize=(20,8))
plt.hist(y_train, bins=len(set(y_train)))
plt.xticks(range(n_class), classes, rotation=90);
```

```python
n_class = len(enc.classes_)
classes = enc.classes_
plt.figure(figsize=(20,8))
plt.hist(y_test, bins=len(set(y_test)))
plt.xticks(range(n_class), classes, rotation=90);
```

```python
if 'net' in locals():
    last_net = net

net = NeuralNet(
    eval_size=0.05,

    layers=[
        ('input', layers.InputLayer),

        ('l1c', Conv2DCCLayer),
        ('l1p', MaxPool2DCCLayer),
        ('l1d', layers.DropoutLayer),

        ('l2c', Conv2DCCLayer),
        ('l2p', MaxPool2DCCLayer),
        ('l2d', layers.DropoutLayer),

        ('l3c', Conv2DCCLayer),
#         ('l3p', MaxPool2DCCLayer),
        ('l3d', layers.DropoutLayer),

        ('l4c', Conv2DCCLayer),
        ('l4p', MaxPool2DCCLayer),
        ('l4d', layers.DropoutLayer),

        ('l5f', layers.DenseLayer),
        ('l5p', layers.FeaturePoolLayer),
        ('l5d', layers.DropoutLayer),

        ('l6f', layers.DenseLayer),
        ('l6p', layers.FeaturePoolLayer),
        ('l6d', layers.DropoutLayer),

#         ('l7f', layers.DenseLayer),
#         ('l7p', layers.FeaturePoolLayer),
#         ('l7d', layers.DropoutLayer),

        ('output', layers.DenseLayer),
    ],

    input_shape=(None, 1, h, w),

    l1c_num_filters=32, l1c_filter_size=(9, 9),
    l1c_W=Orthogonal(),
#     l1c_strides=(2, 2),
    l1p_ds=(2, 2),
    l1d_p=0.2,

    l2c_num_filters=48, l2c_filter_size=(7, 7),
#     l2c_strides=(2, 2),
    l2c_W=Orthogonal(),
    l2p_ds=(2, 2),
    l2d_p=0.2,

    l3c_num_filters=64, l3c_filter_size=(4, 4),
    l3c_W=Orthogonal(),
#     l3c_pad=1,
#     l3p_ds=(2, 2),
    l3d_p=0.2,

    l4c_num_filters=128, l4c_filter_size=(3, 3),
    l4c_W=Orthogonal(),
    l4p_ds=(2, 2),
    l4d_p=0.2,

    l5f_num_units=4096,
    l5f_W=Orthogonal(),
    l5p_ds=2,
    l5d_p=0.5,

    l6f_num_units=4096,
    l6f_W=Orthogonal(),
    l6p_ds=2,
    l6d_p=0.5,

#     l7f_num_units=2048,
#     l7f_W=Orthogonal(),
#     l7p_ds=2,
#     l7d_p=0.5,

    output_num_units=len(set(y)),
    output_nonlinearity=softmax,

    loss=multinomial_nll,

    batch_iterator_train=lasagneutils.MyBatchIterator(batch_size=128),
    batch_iterator_test=lasagneutils.BatchIterator(batch_size=128),

#     batch_iterator_train=BatchIterator(batch_size=256),
#     batch_iterator_test=BatchIterator(batch_size=256),

#     update=nesterov_momentum,
#     update_learning_rate=theano.shared(lasagneutils.float32(0.01)),
#     update_momentum=theano.shared(lasagneutils.float32(0.9)),

    update=rmsprop,
    update_learning_rate=theano.shared(lasagneutils.float32(5e-3)),
    update_rho=0.9,
    update_epsilon=1e-8,

#     update=adadelta,

    regression=False,

    on_epoch_finished=[
        lasagneutils.StepDecay('update_learning_rate', start=5e-3, stop=1e-5),
#         lasagneutils.StepDecay('update_momentum', start=0.9, stop=0.999),
        lasagneutils.EarlyStopping(patience=100)
    ],

    max_epochs=1000,
    verbose=1
)

%time net.fit(X_train, y_train)
```

```python
y_all_pred = net.predict(X)
y_all_pred_proba = net.predict_proba(X)
f1 = f1_score(y_all_pred, y, average=None)
occurances = np.bincount(y)

lscore = utils.multiclass_log_loss(y, y_all_pred_proba)
ascore = net.score(X, y)

print('overall multiclass log loss', lscore)
print('overall accuracy', ascore)
```
