```python
X, y = np.load('data/X_train_48.npy'), np.load('data/y_train_48.npy')
y_enc = utils.load_from_pickle('data/y_train_encoder-2015-02-21-21-04-01.pickle')

X_train, X_test, y_train, y_test = utils.straified_train_test_split(X, y, y=y, test_size=0.25, random_state=42)
```

```python
# net = utils.load_from_pickle('models/net-2.52293672634-2015-02-21-19-11-11.pickle')
# net = utils.load_from_pickle('models/net-net_mar3_specialist_1-0.623163549407-2015-03-04-14-35-06.pickle')
net = utils.load_from_pickle('models/net-net_mar3_specialist_1-0.682265660406-2015-03-05-06-26-17.pickle')
net_layers = net.get_all_layers()
```

### Any dead neurons at conv layers?

```python
for layer in net_layers:
    layer_name = layer.__class__.__name__
    if 'Conv' in layer_name:
        plt.figure()
        utils.vis_layer_weights(layer)
```

### Sanity check dense layer by t-sne

```python
for layer in net_layers:
    layer_name = layer.__class__.__name__
    if 'Dense' in layer_name:
        plt.figure()
        utils.plot_tsne(layer, X_test[:500], y_test[:500])
```
